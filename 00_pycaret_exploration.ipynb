# Procesamiento de datos

Iniciamos convirtiendo los datos categóricos a numéricos.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
categorical_cols = df.select_dtypes(include=['category', 'object']).columns
for col in categorical_cols:
    df[col] = le.fit_transform(df[col].astype(str))

print(df[categorical_cols].nunique())
```

Se busca eliminar las columnas que no aportan información relevante al modelo. Primeramente con la varianza

```python
import numpy as np

var_series = df.var()

#Observamos su distribución

plt.figure(figsize=(10, 6))
plt.hist(var_series, bins=5, edgecolor='black')
plt.title('Distribution of Variance')
plt.xlabel('Variance')
plt.ylabel('Frequency')
plt.grid()
plt.show()

zero_var_cols = var_series[var_series == 0].index

print(f"Columns with zero variance: {zero_var_cols.tolist()}")
```

Se encuentran varias columnas con varianza cero, las cuales no aportan información al modelo. Se eliminan.

```python
df = df.drop(columns=zero_var_cols)
zero_var_cols = df.var()[df.var() == 0].index
print(f"Columns with zero variance after removal: {zero_var_cols.tolist()}")
```

Se revisa la correlación entre las variables para eliminar aquellas que están altamente correlacionadas entre sí.

```python
corr = df.corr()

#Al ser simetrica aplicamos una mascara para visualizarla mejor
mask = np.triu(np.ones_like(corr, dtype=bool))

corr = corr.mask(mask)

plt.figure(figsize=(12, 10))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Correlation Matrix')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.tight_layout()
plt.show()
```

Se observan varias instancias con altas correlaciones, así que se procede a estudiar un treshold para su eliminación.

```python
# Distribución de correlaciones
plt.figure(figsize=(10, 6))
plt.hist(corr.values.flatten(), bins=50, edgecolor='black')
plt.title('Distribution of Correlation Coefficients')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

Así se puede establecer un treshold de 0.9 para eliminar las variables altamente correlacionadas. ya que no incluyen información relevante al modelo y son pocos los casos que se dan.

```python
corr_treshold = 0.9
high_corr_pairs = []
for i in range(len(corr.columns)):
    for j in range(i):
        if abs(corr.iloc[i, j]) > corr_treshold:
            colname = corr.columns[i]
            high_corr_pairs.append((colname, corr.columns[j], corr.iloc[i, j]))

print(f"Number of highly correlated pairs (threshold={corr_treshold}): {len(high_corr_pairs)}")
print("Highly correlated pairs:")
for pair in high_corr_pairs:
    print(f"{pair[0]} and {pair[1]}: {pair[2]:.2f}")
```

Como se puede observar, se encuentran varios pares con valores repetidos en la lista, por lo que se procede a eliminar una de las variables de cada par.

```python
repeated_cols = {}
for pair in high_corr_pairs:
    col1, col2 = pair[0], pair[1]
    if col1 not in repeated_cols:
        repeated_cols[col1] = 1
    else:
        repeated_cols[col1] += 1
    if col2 not in repeated_cols:
        repeated_cols[col2] = 1
    else:
        repeated_cols[col2] += 1

most_repeated_col = sorted(repeated_cols.items(), key=lambda x: x[1], reverse=True)

print("Most repeated columns:")
for col, count in most_repeated_col:
    print(f"{col}: {count} times")
```

```python
# Eliminamos las columnas más repetidas
cols_to_drop = [col for col, count in most_repeated_col if count > 0]
df = df.drop(columns=cols_to_drop)
```

Se revisa de nuevo la correlación entre las variables para ver si se han eliminado las variables altamente correlacionadas.

```python
corr = df.corr()

#Al ser simetrica aplicamos una mascara para visualizarla mejor
mask = np.triu(np.ones_like(corr, dtype=bool))

corr = corr.mask(mask)

plt.figure(figsize=(12, 10))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Correlation Matrix')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.tight_layout()
plt.show()
```

Ahora se reemplazan los valores Nan, pero primero se revisa si existen.

```python
df.isnull().sum().sort_values(ascending=False)
```

En este caso existen en los valores de ECR20 lo que significa que no se detecta actividad por ende se reemplaza por 0.

```python
df = df.fillna(0)
```

```python
print("No missing values left:", df.isnull().sum().sum() == 0)
```

Ahora se buscan distribuciones de datos que puedan ser problematicas para el modelo, iniciando con columnas con distribuciones sesgadas o con pocos valores únicos.

```python
low_variance_cols = []
for col in df.columns:
    if df[col].nunique() < 5 and col != 'FLAG':
        print(f"Column '{col}' has {df[col].nunique()} unique values.")
        low_variance_cols.append(col)
```

Eliminamos las columnas con pocos valores únicos, ya que no aportan información relevante al modelo.

```python
df = df.drop(columns=low_variance_cols)
```

# Preparación de datos

Así se opta por realizar oversampling debido a la diferencia significativa de los datos objetivo, sin embargo como se puede ver en las distribuciones de variación inicial los datos están muy juntos entre ellos lo cual generaría problemas en sus fronteras de clase. Por lo cual se opta por usar SMOTETETomek que combina SMOTE y Tomek links para eliminar los puntos de ruido y mejorar la frontera de decisión.

```python
from imblearn.combine import SMOTETomek
from sklearn.model_selection import train_test_split

X = df.drop(columns=["FLAG"])
y = df["FLAG"]

smote_tomek = SMOTETomek(random_state=23)
print(f"Original dataset shape: {y.value_counts(), X.shape}")
X_resampled, y_resampled = smote_tomek.fit_resample(X, y)
print(f"Resampled dataset shape: {y_resampled.value_counts(), X_resampled.shape}")
```

Separación de test y train

```python
x_train, x_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=23, stratify=y_resampled)
```

## Normalización

Para mejorar el rendimiento del modelo, se normalizan los datos. Se opta por comparar PowerTransformer y MinMaxScaler, ya que PowerTransformer es más robusto a los valores atípicos y puede mejorar la normalización de los datos.

```python
from sklearn.preprocessing import PowerTransformer, MinMaxScaler, StandardScaler, RobustScaler

# Aplicar PowerTransformer
power_transformer = PowerTransformer()

x_data_power_train = power_transformer.fit_transform(x_train)
x_data_power_test = power_transformer.transform(x_test)

# Aplicar MinMaxScaler
min_max_scaler = MinMaxScaler()
x_data_minmax = min_max_scaler.fit_transform(x_train)
x_data_minmax_test = min_max_scaler.transform(x_test)

# Se aplica StandardScaler
standard_scaler = StandardScaler()
x_data_standard_train = standard_scaler.fit_transform(x_train)
x_data_standard_test = standard_scaler.transform(x_test)

#Aplicar RobustScaler
robust_scaler = RobustScaler()
x_data_robust_train = robust_scaler.fit_transform(x_train)
x_data_robust_test = robust_scaler.transform(x_test)
```

Evaluación de la normalización

```python
import scipy.stats as stats
import statsmodels.api as sm

jp_stat_power, jp_p_value_power = stats.jarque_bera(x_data_power_train)
print("Jarque-Bera test for PowerTransformer:" , jp_stat_power, jp_p_value_power)

sm.qqplot(x_data_power_train, line='s')
plt.title('Q-Q Plot for PowerTransformer')
plt.show()

jp_stat_minmax, jp_p_value_minmax = stats.jarque_bera(x_data_minmax)
print("Jarque-Bera test for MinMaxScaler:", jp_stat_minmax, jp_p_value_minmax)

sm.qqplot(x_data_minmax, line='s')
plt.title('Q-Q Plot for MinMaxScaler')
plt.show()

jp_stat_standard, jp_p_value_standard = stats.jarque_bera(x_data_standard_train)
print("Jarque-Bera test for StandardScaler:", jp_stat_standard, jp_p_value_standard)

sm.qqplot(x_data_standard_train, line='s')
plt.title('Q-Q Plot for StandardScaler')
plt.show()

jp_stat_robust, jp_p_value_robust = stats.jarque_bera(x_data_robust_train)
print("Jarque-Bera test for RobustScaler:", jp_stat_robust, jp_p_value_robust)

sm.qqplot(x_data_robust_train, line='s')
plt.title('Q-Q Plot for RobustScaler')
plt.show()
```

En este caso el mejor transformador es el Robust , ya que aunque el Jarques Bera para ambos es significativo, en el qqplot se observa que la normalización de PowerTransformer se ajusta mejor a la línea de referencia, lo que indica una mejor normalidad en los datos. Por lo tanto, se opta por usar PowerTransformer para la normalización de los datos.

```python
x_train = x_data_robust_train
x_test = x_data_robust_test
```

En este caso nos apoyamos en pycaret para el modelado ya que incluye varios modelos de clasificación y permite una fácil comparación entre ellos.

Juntamos de nuevo los datos de entrenamiento y test para usar pycaret ya normalizados.

```python
x_train = pd.DataFrame(x_train, columns=X.columns)
x_test = pd.DataFrame(x_test, columns=X.columns)
df_train = pd.concat([x_train, y_train.reset_index(drop=True)], axis=1)
df_test = pd.concat([x_test, y_test.reset_index(drop=True)], axis=1)
df = pd.concat([df_train, df_test], axis=0)
df = df.reset_index(drop=True)
```
